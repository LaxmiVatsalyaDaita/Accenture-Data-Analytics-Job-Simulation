```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
reactions = pd.read_csv('Reactions.csv')
reaction_types = pd.read_csv('ReactionTypes.csv')
content = pd.read_csv('Content.csv')
```

```python
reaction_types.head()
```

## Choosing Columns

```python
new_content = content[['Content ID', 'User ID', 'Type', 'Category']]
new_reaction = reactions[['Content ID', 'User ID', 'Type']]
new_reaction_types = reaction_types[['Type', 'Sentiment', 'Score']]
```

```python
new_content['Category'] = new_content['Category'].str.lower()
new_content['Category'] = new_content['Category'].str.replace(r'"', "")
```

```python
new_content.nunique()
```

```python
new_content.groupby('Category')["Content ID"].count()
```

```python
new_data = pd.merge(new_reaction, new_content, on='Content ID', how="left")
new_data.head()
```

```python
new_data.rename(columns = {'Type_x':'Reaction_Type', 'User ID_x':'User_ID_(Post_owner)', 'Type_y':'Content_Type'}, inplace = True)
new_data.head()
```

```python
new_data = pd.merge(new_data, new_reaction_types, left_on="Reaction_Type", right_on="Type", how="left")
new_data
```

```python
new_data = new_data.drop('Type', 1)
new_data.head()
```

```python
new_data.isnull().sum()
```

```python
new_data.dropna( inplace=True)
```

```python
new_data.isnull().sum()
```

```python
new_data.columns.values
```

```python
new_data = new_data[['Content ID', 'User_ID_(Post_owner)', 'Reaction_Type', 
       'Content_Type', 'Category', 'Sentiment', 'Score']]
```

```python
new_data.to_csv('DataCleaned.csv', index=False)
```
